```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("sales.xlsx") #reading excel file from the local machine.
```

```python
df.head() 
```

```python
df.shape #cheaking the shape of the data frame. There are 51290 row and 21 column.
```

```python
df.isnull().sum() #cheaking the null value in the data so that we cal do the correct analysis.we can see there are not any null value in this data frame.
```

# The 15 most selling products by quantity.


```python
most_selling_product=df.groupby('product_name').agg({'quantity':'sum'}).sort_values('quantity',ascending=False)[:15]
most_selling_product
```

```python
sn.barplot(x='product_name',y='quantity',data=most_selling_product.reset_index())
plt.xticks(rotation='vertical')
```

# The Top 15 products by sales.

```python
sales_product=df.groupby('product_name').agg({'sales':sum}).sort_values('sales',ascending=False)[:15]
sales_product
```

```python
sn.barplot(x='product_name',y='sales',data=sales_product.reset_index())
plt.xticks(rotation='vertical')
```

# The top 15 product that are most profitable sales.

```python
profitable_product=df.groupby('product_name').agg({'profit': sum}).sort_values('profit',ascending=False)[:15]
profitable_product
```

```python
sn.barplot(x='product_name',y='profit',data=profitable_product.reset_index())
plt.yticks(rotation='vertical')
plt.xticks(rotation='vertical')
```

# Top 15 product sales by category.

```python
df.dtypes #cheaking the datatypes.
```

```python
df=df.drop(['order_date','ship_date'],axis=1) #datetime64 type does not support sum operations occure so that i drop this column to group the product with category.
```

```python
df.groupby(['category','sales']).sum()
```

# The most profitable category.

```python
profitable_category=pd.DataFrame(df.groupby(['category']).sum()['profit'])
profitable_category.sort_values(by=['profit'],inplace=True,ascending=False)
profitable_category[:15]
```

```python
sn.barplot(x='category',y='profit',data=profitable_category.reset_index())
```

```python
df.groupby(['category','sub_category']).sum()
```

# Which customer segments are the most profitable?

```python
selling_sub_category=df.groupby(['category','sub_category'],as_index=False)["quantity"].count()
selling_sub_category
```

```python
profit_by_segement=pd.DataFrame(df.groupby(['segment']).sum()['profit'])
profit_by_segement.sort_values(by=['profit'],inplace = True, ascending=False)
profit_by_segement
```

```python
sn.barplot(x='segment',y='profit',data=profit_by_segement.reset_index())
```

# What shipping modes sold the most products?

```python
shipping_mode=pd.DataFrame(df.groupby(['ship_mode']).sum()['sales'])
shipping_mode.sort_values(by=['sales'],inplace = True, ascending=False)
shipping_mode
```

```python
sn.barplot(x='ship_mode',y='sales',data=shipping_mode.reset_index())
```

# Which are the Top 10 country by sales?

```python
country_by_sales=pd.DataFrame(df.groupby(['country']).sum()['sales'])
country_by_sales.sort_values(by='sales',inplace=True,ascending=False)
country_by_sales.head(10)
```

```python
countries = country_by_sales.sort_values(by = 'sales',ascending = False)[1:11]
countries
```

```python
countries['sales'].plot(kind='pie',autopct='%1.1f%%',figsize=(14,7))
plt.title('10 Countries with most sales')
plt.show()
```
